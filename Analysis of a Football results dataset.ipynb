```python
import pandas as pd

```

```python
df = pd.read_csv('results.csv')
```

```python
df.head(10)
```

```python
# Create a column called 'home_team_won'where the home score is higher than the away score

df['home_team_won'] = df['home_score']> df['away_score']
```

```python
df
```

```python
# Create a column called 'away_team_won'where the home score is lower than the away score

df['away_team_won'] = df['home_score']< df['away_score']
```

```python
df
```

```python
# Create a second dataframe where either 'home_team_won' or 'away_team_won' is True (thus avoiding ties)

df2 = df[(df['home_team_won'] == True) | (df['away_team_won'] == True)]
```

```python
df2
```

```python
# Create an empty dictionary team_wins


team_wins = {}

# Define the function add_goals_to_team

def add_goals_to_team(team, goals):
    if team in team_wins:
        team_wins[team] += goals
    else:
        team_wins[team] = goals

# Loop over the df2

for index,row in df2.iterrows():
    
    if row['home_team_won'] == True:
        add_goals_to_team(row['home_team'], int(row['home_score']))
        
    else:
        add_goals_to_team(row['away_team'], int(row['away_score']))
        
        
    
```

```python
team_wins
```

```python
# Create a dataframe from the dictionary team_wins

df3 = pd.DataFrame.from_dict(team_wins, orient= 'index', columns = ['wins'] )
```
